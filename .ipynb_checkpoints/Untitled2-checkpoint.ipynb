```python
#NETWORK MEDIATED SEPARATION
import random

sz_list = list(range(len(on_list)))
k = sz_k
r = sz_r
vth = sz_vth

n_on = 5

for c,n_add in enumerate(diff_list):
#Reliability - distance between same input
    stim_time = 401
    n_input = N

    for i in range(1):
        norm_vec = []
        on_ind = random.sample(list(np.arange(0, n_input)), n_on)
        #Define synaptic connections from input - same across each stimulus size across network conditions
        input_A = np.zeros((N,N))
        np.fill_diagonal(input_A, 1)
        input_W = np.ones(input_A.shape)

        bind1, bind_inp1, bind_og = stim_net(sim_time, stim_time, n_on, on_ind, input_A, input_W, k, vth, r, s, divisor, soften, N, dist, v_rest, t_syn_del, tau_l, N_e, lam, w_e)



        on_ind = random.sample(list(np.arange(0, n_input)), n_on + n_add)
        #Define synaptic connections from input - same across each stimulus size across network conditions
        input_A = np.zeros((N,N))
        np.fill_diagonal(input_A, 1)
        input_W = np.ones(input_A.shape)

        bind2, bind_inp2, bind_og = stim_net(sim_time, stim_time, n_on + n_add, on_ind, input_A, input_W, k, vth, r, s, divisor, soften, N, dist, v_rest, t_syn_del, tau_l, N_e, lam, w_e)
        norm_ed = np.linalg.norm(bind_inp1[:,stim_time] - bind_inp2[:,stim_time]) - np.linalg.norm(bind1[:,stim_time] - bind2[:,stim_time])
        norm_vec = np.append(norm_vec, norm_ed)
    sz_list[c] = norm_vec
    print(n_on + n_add)
```
